#Pequeño proyecto sobre un ejercicio que pide investigar si con el correr del paso del tiempo, las peliculas de netflix han acortado la duración de sus peliculas con el correr del tiempo.

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv("netflix_data.csv",index_col = 0)
```

```python
netflix_df
```

Filtramos los datos para eliminar los programas de TV y almacenarlos en un nuevo dataset:

```python
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
netflix_subset
```

```python
netflix_movies = netflix_subset[['title','country','genre','release_year','duration']]
netflix_movies
```

Filtramos las peliculas que duren menos de minutos y las guardamos en otro dataset llamado 'short_movies':

```python
short_movies = netflix_movies[netflix_movies['duration']<60]
short_movies
```

```python
print(short_movies['genre'].unique().tolist())
```

```python
print("Cantidad de peliculas cortas que hay por genero: ")
print("{}".format(short_movies['genre'].value_counts()))
```

Observamos que los generos de peliculas cortas son: Documentales, infantiles, stand-up, sin categoria, comedias, dramas, peliculas clasicas, accion, peliculas del cine independientes, anime, musicales y peliculas de horror. Esto tiene sentido ya que para algunas categorias de estas peliculas (Como las de cine independiente, horror, comedia, dramas o infantiles) son normalmente peliculas de bajo presupuesto y por otro lado, tenes categorias que ni siquiera son peliculas como documentales, stand-up o animes donde estos ultimos no tienen la necesidad de contar historias tan largas (O como el caso de los documentales que ya de por si, son extremadamente caros de filmar).

##Recorremos las filas del dataset de netflix_movies para asignar colores a cuatro grupos de géneros ('children', 'documentaries', 'stand-up' y 'otros'):

```python
# Almacena los resultados en una lista
color_rojo = []
color_verde = []
color_azul = []
color_negro = []
for indice,fila in netflix_movies.iterrows():
  if (fila['genre'] == 'Documentaries'):
    color_rojo.append('red')
  elif (fila['genre'] == 'Children'):
    color_verde.append('green')
  elif (fila['genre'] == 'Stand-Up'):
    color_azul.append('blue')
  else:
    color_negro.append('black')
```

```python
aux_netflix_red = netflix_movies[netflix_movies['genre'] == 'Documentaries']
aux_netflix_green = netflix_movies[netflix_movies['genre'] == 'Children']
aux_netflix_blue = netflix_movies[netflix_movies['genre'] == 'Stand-Up']
aux_netflix_black = netflix_movies[(netflix_movies['genre'] != 'Documentaries') & (netflix_movies['genre'] != 'Children') & (netflix_movies['genre'] != 'Stand-Up')]
```

Graficamos en un gráfico de dispersión:

Para el género de documentales:

```python
plt.scatter(aux_netflix_red['release_year'],aux_netflix_red['duration'], color = color_rojo)
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duración (En minutos)")
plt.title("Duración de la película del genero documental por año de estreno")
plt.show()
```

Para el género de infantiles:

```python
plt.scatter(aux_netflix_green['release_year'],aux_netflix_green['duration'], color = color_verde)
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duración (En minutos)")
plt.title("Duración de la película del genero infantil por año de estreno")
plt.show()
```

Para el genero de Stand-Up:

```python
plt.scatter(aux_netflix_blue['release_year'],aux_netflix_blue['duration'], color = color_azul)
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duración (En minutos)")
plt.title("Duración de la película del genero Stand-Up por año de estreno")
plt.show()
```

Para otros generos:

```python
plt.scatter(aux_netflix_black['release_year'],aux_netflix_black['duration'], color = color_negro)
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duración (En minutos)")
plt.title("Duración de la película de otros generos por año de estreno")
plt.show()
```
